```python
%matplotlib qt

import matplotlib.pyplot as plt
import numpy as np
from cedalion.sigproc.BVPWaveformAnalysis_v12 import peakseek

t_y = np.arange(0, 30*np.pi, np.pi/32)
y = np.sin(t_y)

t_x = np.arange(np.pi/2, 30*np.pi, 2*np.pi)
x = np.sin(t_x)

max_idx, max_val = peakseek(y)

plt.plot(t_y,y)
plt.plot(t_x, x, '+', color='b')
plt.plot(t_y[max_idx], max_val, '+', color='r')
```
